```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[7,2],[-7,5]])
b = np.array([-5,12])
```

```python
# np의 모듈인 solve <- 해를 구합니다
x = np.linalg.solve(A,b)
x
```

```python
# 해 벡터인 x를 그래프로 표현하기 위해 x1,y1으로 표기합니다.
x1, y1 = x
print(x1,y1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
```

```python
# 선형 방정식을 직선으로 그립니다
a1 = A[:,0]
b1 = A[:,1]
c1 = -b
print(a1)
print(b1)
print(c1)
```

```python
for c1,c2,c3 in zip(a1,b1,b):
    x = np.linspace(-7,7,100)
    y = (c3-c1*x)/c2
    ax.plot(x,y,color="black")
```

```python
ax.plot(x1,y1,"ro")
```

```python
ax.axis([-7,7,-7,7])
ax.set_xticks(range(-7,7))
ax.set_yticks(range(-7,7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('auto',adjustable = 'box')
```
